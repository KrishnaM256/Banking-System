# Using statsmodels

```python
import statsmodels.api as sm
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('./data_files/House_Price.csv',header=0)
```

```python
df.head()
```

```python
X = sm.add_constant(df['room_num'])
```

```python
lm = sm.OLS(df['price'],X).fit()
```

```python
lm.summary()
```

```python
sns.jointplot(x='room_num',y='price',data=df)
```

```python
sns.jointplot(x='room_num',y='price',data=df,kind='reg')
```

# Using sklearn

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
y = df['price']
```

```python
X = df[['room_num']]
```

```python
lm2 = LinearRegression()
```

```python
lm2.fit(X,y)
```

```python
print(lm2.coef_,lm2.intercept_)
```

```python
lm2.predict(X)
```

```python
plt.scatter(x=df['room_num'],y=df['price'])
```
