```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
```

```python
df = pd.read_csv('./data_files/House_Price.csv',header=0)
```

```python
df.head()
```

```python
X_multi = df.drop("price",axis=1)
```

```python
X_multi
```

```python
X_multi_const = sm.add_constant(X_multi)
```

```python
X_multi_const
```

```python
df2 = pd.get_dummies(df,drop_first=True) + 0
df2
```

```python
df2.head()
df2.replace([np.inf, -np.inf], np.nan, inplace=True)
df2.dropna(inplace=True)
```

```python
X_multi = df2.drop('price',axis=1)
```

```python
X_multi
```

```python
X_multi_const = sm.add_constant(X_multi)
```

```python
X_multi_const
```

```python
y_multi = df2['price']
```

```python
y_multi
```

```python
multi_lm = sm.OLS(y_multi,X_multi_const).fit()
```

```python
multi_lm
```

```python
multi_lm.summary()
```
